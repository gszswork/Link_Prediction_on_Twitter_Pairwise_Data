```python
import numpy as np
import networkx as nx
import pickle
from tqdm import tqdm

```

```python
def generate_features(data, name, DG, UG):    
    HAA = []
    HJC = []
    HRA = []
    SD = []
    for e in tqdm(data):
        e = e[0]
        if not DG.has_node(e[0]):
            DG.add_node(e[0])
            UDG.add_node(e[0])
        if not DG.has_node(e[1]):
            DG.add_node(e[1])
            UDG.add_node(e[1])
        AA = nx.adamic_adar_index(UDG, [e])
        JC = nx.jaccard_coefficient(UDG, [e])
        RA = nx.resource_allocation_index(UDG, [e])
        spec_diff = DG.in_degree(e[1]) - DG.in_degree(e[0])  # specificity_difference
        SD.append(spec_diff)
        try:
            for u, v, p in AA:
                HAA.append(p)
        except ZeroDivisionError:
                HAA.append(0)
                pass

        try:
            for u, v, p in JC:
                HJC.append(p)
        except ZeroDivisionError:
                HJC.append(0)
                pass

        try:
            for u, v, p in RA:
                HRA.append(p)
        except ZeroDivisionError:
                HRA.append(0)
                pass



    '''Store the feature scores'''
    np.savez_compressed(name, HAA=HAA, HJC=HJC, HRA=HRA, SD=SD)
    print('all finished! Features saved')
```

```python
generate_features(pos_data, "new_positive_original", DG, UDG)
generate_features(neg_data, "new_negative_original", DG, UDG)
```
