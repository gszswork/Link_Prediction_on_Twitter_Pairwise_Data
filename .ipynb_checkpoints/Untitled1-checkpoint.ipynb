```python
import pickle 


dirname = 'data/'
def save_obj(obj, name ):
    with open(dirname+ name + '.pkl', 'wb') as f:
        pickle.dump(obj, f)

def load_obj(name ):
    with open( dirname + name + '.pkl', 'rb') as f:
        return pickle.load(f)
```

```python
training_edges = load_obj('training_edges')
```

```python
save_obj(training_edges, 'training_edges')
```

```python
pre_features = load_obj('pre_features_v2')
save_obj(pre_features, 'pre_features_v2')
```
